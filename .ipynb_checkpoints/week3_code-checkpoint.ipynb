```python
#Importing necessay libraries and modules
import numpy as np
import scipy.special as sp 
from pylab import * 
from scipy import linalg
FILENAME = "fitting.dat"
```

```python
N=101                           # no of data points
k=9                             # no of sets of data with varying noise

# generate the data points and add noise
t=linspace(0,10,N)              # t vector
y=1.05*sp.jn(2,t)-0.105*t       # f(t) vector
Y=meshgrid(y,ones(k),indexing='ij')[0] # make k copies
scl=logspace(-1,-3,k)           # noise stdev
n=dot(randn(N,k),diag(scl))     # generate k vectors
yy=Y+n                          # add noise to signal
savetxt("fitting.dat",c_[t,yy]) # write out matrix to file
```

```python
data = np.loadtxt(FILENAME) #Loading data
```

```python
sigma = list(map(lambda x: float("{0:.3f}".format(x)), scl))

# shadow plot
figure(figsize=(8,6))
plot(t,yy)
legend(sigma)
xlabel(r'$t$',size=20)
ylabel(r'$f(t)+n$',size=20)
title(r'Plot of the data to be fitted')
grid(True)
show()
```

```python
def g(t, A = 1.05, B = -0.105):
    return A * sp.jn(2, t) + B*t


y_t = list(map(g, t))

figure(figsize=(8,6))
plot(t,yy, label = sigma)
plot(t, y_t, label = "True Value")
legend()
xlabel(r'$t$',size=20)
ylabel(r'$f(t)+n$',size=20)
title(r'Plot of the data to be fitted')
grid(True)
show()
```

```python
figure(figsize=(8,6))
plot(t, y_t)
errorbar(t[::5],yy[::5,0],sigma[0],fmt="ro")
legend(["True Value", "Error Bars"])
xlabel(r'$t$',size=20)
ylabel(r'$f(t)+n$',size=20)
title(r'True value and errorbars')
grid(True)
show()
```

```python
A, B = 1.05, -0.105
Js = np.apply_along_axis(lambda x: sp.jn(2, x), axis = 0, arr = t)
M = c_[Js, t]
Gs = np.apply_along_axis(g, axis = 0, arr = t).reshape(-1,1)
if np.allclose(np.matmul(M, np.array([A, B]).reshape(2,1)), Gs):
    print("The vectors are same")
else:
    assert False, "The vectors are not equal"
```

```python
A = np.linspace(0, 2, 21).tolist()
B = np.linspace(-0.2, 0, 21).tolist()
e = np.zeros((2,21,21))
print(t.shape)
for _ in range(2):
    for i in range(21):
        for j in range(21):
            Gs = np.apply_along_axis(lambda x: g(x,A[i],B[j]), axis = 0, arr = t)
            e[_][i][j] += 1/101*np.sum((yy[:,_] - Gs)**2, axis = 0)
```

```python
figure(figsize=(8,6))
contour(A,B,e[0])
contour(A,B,e[1])
xlabel(r'$A$',size=20)
ylabel(r'$B$',size=20)
title(r'Error contours')
grid(True)
show()
```

```python
Gs = np.apply_along_axis(g, axis = 0, arr = t).reshape(-1,1)
GroundTruth = linalg.lstsq(M, Gs)[0]
x = np.zeros((yy.shape[1], 2))
for i in range(yy.shape[1]):
    x[i] = linalg.lstsq(M, yy[:,i])[0]
```

```python
err = np.zeros((yy.shape[1]))
for i in range(yy.shape[1]):
    err[i] += ((x[i][0] - GroundTruth[0])**2 + (x[i][1] - GroundTruth[1])**2)**0.5

figure(figsize=(8,6))
plot(sigma,err, label = "Line1") 
xlabel(r'$\sigma$',size=20)
ylabel(r'$Error(Euclidian)$',size=20)
legend()
title(r'Error vs Sigma')
grid(True)
show()
```

```python
figure(figsize=(8,6))
plot(sigma,err, label = "Line1") 
xlabel(r'$\sigma$',size=20)
ylabel(r'$Error(Euclidian)$',size=20)
yscale("log")
xscale("log")
legend()
title(r'Error vs Sigma')
grid(True)
show()
```
